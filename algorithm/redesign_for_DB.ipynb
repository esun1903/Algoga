```python
import pandas as pd
import numpy as np
```

```python
algo = pd.read_csv("raw_data/algorithms.csv")

algo.info()
algo.head()
```

```python
algo.drop(['algo_name_en', 'algo_url', 'algo_cnt'], axis = 1, inplace = True)
algo = algo.rename({'algo_name_kr' : 'algo'}, axis = 1)
algo['seq'] = list(range(len(algo)))

algo = algo[['seq', 'algo_id', 'algo']]

algo.info()
algo.head(172)
```

```python
algo.to_csv("DB_data/type.csv", index = False, encoding='utf-8-sig')
pd.to_pickle(algo, "DB_data/type.pkl")
```

```python
pro_with_algo = pd.read_csv("raw_data/all_problem_with_algorithm.csv")

pro_with_algo.info()
pro_with_algo.head()
```

```python
pro_with_algo.drop(['name', 'algo'], axis = 1, inplace = True)

pro_with_algo = pro_with_algo.rename({'id' : 'number'}, axis = 1)

pro_with_algo.info()
pro_with_algo.head()
```

```python
problems = pd.read_csv("final_data/CSV/recommend_problem_infos.csv")

problems.info()
problems.head()
```

```python
problem2idx = {}
for seq, number in zip(problems['seq'], problems['number']) :
    problem2idx[number] = seq;
len(problem2idx)
```

```python
algo2idx = {}
for seq, algo_id in zip(algo['seq'], algo['algo_id']) :
    algo2idx[algo_id] = seq;
    
len(algo2idx)
```

```python
problems = problems['number'].unique()
```

```python
pro_with_algo = pro_with_algo[pro_with_algo['number'].isin(problems)]
```

```python
pro_with_algo.info()
pro_with_algo.head()
```

```python
len(pro_with_algo['number'].unique())
```

```python
pro_with_algo['problem_seq'] = list(map(lambda x : problem2idx[x], pro_with_algo['number']))
pro_with_algo['algo_seq'] = list(map(lambda x : algo2idx[x], pro_with_algo['algo_id']))
pro_with_algo['seq'] = range(len(pro_with_algo))
```

```python
pro_with_algo.info()
pro_with_algo.head()
```

```python
pro_with_algo = pro_with_algo[['seq', 'problem_seq', 'algo_seq', 'number', 'algo_id']]

pro_with_algo.info()
pro_with_algo.head()
```

```python
pro_with_algo.to_csv("DB_data/type_of_problems.csv", index = False, encoding='utf-8-sig')
pd.to_pickle(pro_with_algo, "DB_data/type_of_problems.pkl")
```

```python
problems = pd.read_csv("final_data/CSV/recommend_problem_infos.csv")

problems_info = problems[['seq', 'number', 'information']]

problems_info.info()
problems_info.head()
```

```python
problems_info['information_list'] = problems_info['information'].str.split(",")

problems_info.info()
problems_info.head()
```

```python
problems_info = problems_info.explode('information_list')

problems_info.info()
problems_info.head()
```

```python
problems_info.drop('information', axis = 1, inplace = True)
```

```python
problems_info = problems_info.rename({'information_list' : 'information'}, axis = 1)
```

```python
problems_info = problems_info.rename({'seq' : 'problem_seq'}, axis = 1)
```

```python
problems_info.info()
problems_info.head()
```

```python
information_list = list(problems_info['information'].unique())

information2idx = dict(zip(information_list, range(len(information_list))))

information_df = pd.DataFrame({
    'seq' : list(information2idx.values()),
    'name' : list(information2idx.keys())
})
```

```python
information_df.info()
information_df.head()
```

```python
information_df.to_csv("DB_data/information.csv", index = False, encoding='utf-8-sig')
pd.to_pickle(information_df, "DB_data/information.pkl")
```
