```python
# 텍스트 형태로 제출하는 문제들은 제외
text_problems = ['1237', '2555', '5620', '9073', '9987', '9995', '9999', '10212',
                '10831', '10943', '10944', '10945', '10946', '10947', '10948',
                '11506', '12091', '12092', '12093', '12096', '13062', '14406',
                '14443', '14918', '15547', '15631', '15635', '15636', '15637',
                '15639', '15641', '15643', '15802', '15891', '15913', '15915',
                '16075', '16076', '16647', '17107', '17108', '17110', '17111',
                '17112', '17113', '17116', '17467', '17468', '18169', '18351',
                '18822', '18823', '18824', '18825', '18826', '18828', '18829',
                '19617', '19999']

text_problems = list(map(lambda x : int(x), text_problems))
```

```python
# 텍스트형태로 제출하는 문제는 제외
baekjoon = baekjoon[~baekjoon['id'].isin(text_problems)]
solved = solved[~solved['id'].isin(text_problems)]
baekjoon_with_algorithm = baekjoon_with_algorithm[~baekjoon_with_algorithm['id'].isin(text_problems)]
```

```python
algorithm_problems = baekjoon_with_algorithm['id'].unique()
len(algorithm_problems)
```

```python
problem_seqs = []
problem_numbers = []
problem_names = []
algorithms = []
algorithm_ids = []

for index, problem_id in enumerate(algorithm_problems) :
    
    name = baekjoon_with_algorithm[baekjoon_with_algorithm['id'] == problem_id]['name'].unique()[0]
    algorithm = list(baekjoon_with_algorithm[baekjoon_with_algorithm['id'] == problem_id]['algo'])
    algorithm_id = list(map(str, baekjoon_with_algorithm[baekjoon_with_algorithm['id'] == problem_id]['algo_id']))
    
    problem_seqs.append(index)
    problem_numbers.append(problem_id)
    problem_names.append(name)
    algorithms.append(",".join(algorithm))
    algorithm_ids.append(",".join(algorithm_id))
```

```python
len(problem_seqs), len(problem_numbers), len(problem_names), len(algorithms), len(algorithm_ids)
```

```python
recommend_problem_df = pd.DataFrame({
    'seq' : problem_seqs,
    'number' : problem_numbers,
    'name' : problem_names,
    'algorithm' : algorithms,
    'algorithm_id' : algorithm_ids,
})
```
