```python
for index, row in baekjoon_with_algorithm.iterrows() :
    problem_id = row['id']
    algo = row['algo']
    
    baekjoon.loc[baekjoon['id'] == problem_id, algo] = 1
```

```python
baekjoon.info()
baekjoon.head()
```

```python
baekjoon['algo_cnt'] = baekjoon.iloc[:,6:].sum(axis=1)

baekjoon.info()
baekjoon.head()
```

```python
recommend_problems = baekjoon.loc[baekjoon['algo_cnt'] != 0, :]
recommend_problems.info()
recommend_problems.head()
```
